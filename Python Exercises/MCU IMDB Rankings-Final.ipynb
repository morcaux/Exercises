```python
#Importing Data Sets
import pandas as pd
import numpy  as np
import matplotlib.pyplot as plt
df = pd.read_csv("MCU_IMDB.csv")
df.head()
```

```python
##Shifting the Title column as the Leftmost column
first_column = df.pop('Title')
df.insert(0, 'Title', first_column)
print("Shifted Column")
df.head()
del df["Pageurl"]
del df["Uniq Id"]
print("New Dataframe")
df.head()
```

```python
#Formatting dollar signs and changing data types
df['Collection'] = df['Collection'].replace('[\$\,\.]',"",regex=True).astype(float)
df['Budget'] = df['Budget'].replace('[\$\,\.]',"",regex=True).astype(float)
#Checking datatypes 2nd time
df.dtypes
df.head()
```

```python
#Checking datatypes
df.dtypes
```

```python
#Exploratory Data Analysis and Visualization
```

```python
##Top Movies by Box Office Gross
boxoffice = df[["Title","Collection","Budget"]]
Earnings = boxoffice.sort_values(by = ['Collection'], ascending = False)
Earnings.head()
```

```python
#Observation: All of the Avenger movies are in the top 5 Grossing Movies of the MCU
```

```python
##Bar Plot Box Office Gross vs Budget 
Earnings.iloc[0:5].plot(x="Title", kind="bar" ) 
plt.show()
```

```python
##Top Movies by Box Office Net
###The Budget will be substracted to the Collection Column to get the Net earnings of each movies
```

```python
Earnings['Net'] = Earnings['Collection'] - Earnings['Budget']
```

```python
Earnings.head()
```

```python
Moolah = Earnings.sort_values(by=['Net'], ascending=False)
Moolah.head()
```

```python
##Observation: The Top Five highest Grossing movies also resulted with the highest net.
```

```python
##The movies with the least Net Earnings
NoMoolah = Earnings.sort_values(by=['Net'], ascending=True)
NoMoolah.head()
```

```python
ratingx = df[["Title", "Review Rating", "Collection"]]
MoolahRatings = ratingx.sort_values(by=['Review Rating'], ascending=False)
```

```python
#Movie Review Ratings and Box Office
MoolahRatings.head()
```

```python
MoolahRatings.plot("Review Rating", "Collection", kind = "scatter",title = "Ratings vs Gross",figsize=(10,5))
plt.show()
```

```python
MoolahRatings['Review Rating'].hist(bins=20, range=(0,10))
plt.show()
```

```python
MoolahRatings['Collection'].hist(bins=20, range=(0,3000000000))
plt.show()
```

```python
##As you see, the Box office gross doesn't necessarily correlate to Review Ratings. 
### It could be inferred that the Avengers Movies are all box office successes. And also with the highest Review Ratings among its MCU counterparts
### For the exception of Age of Ultron, though it gained money in the box office, it didn't garner much success from the Critics. 
```
